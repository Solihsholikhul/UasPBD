```python
df = pd.DataFrame(index=['Agus', 'Budi', 'Bagas', 'Kopek', 'Rio'], 
                  data={ 'Jambu':[8, 30, 12, 2, 10],
                        'Durian':[35, 40, 25, 19, 33]})
df
```

Penjelasan : Dta akan keluar sesuai dengan indexnya masing-masing contoh index 0 dimiliki oleh Agus, 8, 35. dst  

```python
color = ['.2', '.7']
df.plot(kind='bar', color=color, figsize=(16,4))
```

Penjelasan : Jadi Agus lebih banyak memiliki Durian dari pada jambu, dari keseluruhan lebih banyak memiliki durian dari pada jambu karena lagi musim durian.

```python
df.plot(kind='kde', color=color, figsize=(16,4))
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(16,4))
fig.suptitle('Two Variable Plots', size=20, y=1.02)
df.plot(kind='line', color=color, ax=ax1, title='Line plot')
df.plot(x='Jambu', y='Durian', kind='scatter', color=color, 
        ax=ax2, title='Scatterplot')
df.plot(kind='bar', color=color, ax=ax3, title='Bar plot')
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(16,4))
fig.suptitle('One Variable Plots', size=20, y=1.02)
df.plot(kind='kde', color=color, ax=ax1, title='KDE plot')
df.plot(kind='box', ax=ax2, title='Boxplot')
df.plot(kind='hist', color=color, ax=ax3, title='Histogram')
```

# Visualizing the flights dataset

Sumber dari dataset (URL) :
https://www.kaggle.com/usdot/flight-delays/data?select=flights.csv

```python
flights = pd.read_csv('flights.csv')
flights.head()
```

```python
flights['ARRIVAL_DELAY'] = flights['AIRLINE_DELAY'].ge(15).astype(int)
cols = ['DIVERTED', 'CANCELLED']
flights['ON_TIME'] = 1 - flights[cols].any(axis=1)

cols.append('ON_TIME')
status = flights[cols].sum()
status
```

```python
fig, ax_array = plt.subplots(2, 3, figsize=(18,8))
(ax1, ax2, ax3), (ax4, ax5, ax6) = ax_array
fig.suptitle('2015 US Flights - Univariate Summary', size=20)

ac = flights['AIRLINE'].value_counts()
ac.plot(kind='barh', ax=ax1, title ='Airline')

oc = flights['FLIGHT_NUMBER'].value_counts()
oc.plot(kind='bar', ax=ax2, rot=0, title='TAIL_NUMBER')

dc = flights['ORIGIN_AIRPORT'].value_counts().head(10)
dc.plot(kind='bar', ax=ax3, rot=0, title='Destination_AIRPORT')
```
